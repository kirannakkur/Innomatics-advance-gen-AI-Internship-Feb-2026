```python
#Problem 1: Mobile Recharge Validation System
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("‚ùå Recharge amount must be at least ‚Çπ50.")
        return False

    if amount not in valid_plans:
        print("‚ùå Invalid plan selected.")
        print("Available plans: 199, 299, 399, 599")
        return False

    return True


# Main program
while True:
    try:
        amount = int(input("Enter recharge amount: ‚Çπ"))

        if validate_recharge(amount):
            print("‚úÖ Recharge successful!")
            break   # exit loop if valid
        else:
            print("üîÅ Please try again.\n")

    except ValueError:
        print("‚ùå Please enter a valid numeric amount.\n")
```

```python
#Problem 2: Inventory Reorder Alert System
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product} ‚Üí ‚ö† Reorder Alert (Stock: {quantity})")
        else:
            print(f"{product} ‚Üí ‚úÖ Stock OK (Stock: {quantity})")


# Sample Inventory Data
inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 5,
    "Soap": 18,
    "Milk Powder": 12
}

# Function Call
check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System
def process_result(marks):
    total = 0

    # Calculate total using loop
    for mark in marks:
        total += mark

    average = total / len(marks)

    # Decide result
    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"


# Sample Input
student_marks = [60, 45, 70, 50, 40]

# Function Call
avg, result = process_result(student_marks)

print("Average Marks:", avg)
print("Result:", result)
```

```python
#Problem 4: Cab Fare Estimator with Retry Option
def calculate_fare(distance, is_peak):
    base_fare = 50
    rate_per_km = 12

    fare = base_fare + (distance * rate_per_km)

    if is_peak:
        fare += fare * 0.25   # 25% extra

    return fare


# Main Program with Retry
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak_input = input("Is it peak hour? (yes/no): ").lower()

        is_peak = True if peak_input == "yes" else False

        total_fare = calculate_fare(distance, is_peak)

        print(f"Total Fare: ‚Çπ{total_fare:.2f}")

        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            print("Thank you for using Cab Fare Estimator!")
            break

    except ValueError:
        print("‚ùå Please enter valid numeric distance.\n")
```

```python
#Problem 5: Employee Attendance Eligibility Checker
def check_eligibility(attendance_list):
    present_count = 0

    # Count present days using loop
    for day in attendance_list:
        if day == "P":
            present_count += 1

    total_days = len(attendance_list)
    percentage = (present_count / total_days) * 100

    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"


# Sample Input
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

# Function Call
percent, status = check_eligibility(attendance)

print("Attendance Percentage:", percent)
print("Status:", status)
```

```python
#Problem 6: Password Strength Checker
def check_password_strength(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Rule 1: Minimum length
    if len(password) < 8:
        return False

    # Loop through password
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    # Final condition
    if has_digit and has_special:
        return True
    else:
        return False


# Main Program with Retry
while True:
    user_password = input("Enter your password: ")

    if check_password_strength(user_password):
        print("Strong Password")
        break
    else:
        print("Weak Password")
        print("Password must:")
        print("- Be at least 8 characters")
        print("- Contain at least one digit")
        print("- Contain at least one special character (@#$)")
        print("üîÅ Please re-enter.\n")
```
