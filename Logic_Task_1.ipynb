```python
#1. User Login Check /

# Given credentials
username = "admin"
password = "1234"

# Take input from User
input_username = input("Enter username: ")
input_password = input("Enter password: ")

# Login check
if input_username == username and input_password == password:
  # If both username and password match
    print("Login Successful ✔")
else:
  # If either username or password is incorrect
    print("Invalid Credentials")
```

```python
#2. Pass / Fail Analyzer /

# List of student marks
marks = [45, 78, 90, 33, 60]

# Initialize counters
pass_count = 0
fail_count = 0

# Analyze student marks
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Displaying student results
print("Total Students:", len(marks))
print("Number of Students Passed:", pass_count)
print("Number of Students Failed:", fail_count)
```

```python
#3. Simple Data Cleaner /

# List of user names
names = [" Alice ", "bob", " CHARLIE "]

# Initialize cleaned list
cleaned_names = []

# Clean each name
for name in names:
    cleaned_name = name.strip().lower()  # Remove spaces and convert to lowercase
    cleaned_names.append(cleaned_name)

# Display cleaned list
print("Cleaned Names:", cleaned_names)
```

```python
#4. Message Length Analyzer /

# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze each message
for message in messages:
    length = len(message)  # Calculate message length

    print("Message:", message)
    print("Length:", length)

    # Check if message is longer than 10 characters
    if length > 10:
        print("Status: Message is longer than 10 characters")

    print()  # Blank line for better readability
```

```python
#5. Error Message Detector /

# List of system log entries
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Initialize error counter
error_count = 0

# Check each log entry
for log in logs:
    if log == "ERROR":
        error_count += 1  # Increment counter if log is ERROR

# Display total error count
print("Total ERROR entries:", error_count)
```
